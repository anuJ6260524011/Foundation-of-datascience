```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

print("\n### Experiment 1: Working with NumPy Arrays ###")
arr = np.arange(1, 101)
print(arr)
print(arr.shape)
print(arr.reshape(10, 10))
print(np.sum(arr))
print(np.mean(arr))
print(np.std(arr))
print(np.max(arr))
print(np.min(arr))
print(np.cumsum(arr))
print(np.cumprod(arr))
print(np.percentile(arr, [25, 50, 75]))
print(np.linalg.norm(arr))

print("\n### Experiment 2: Working with Pandas DataFrames ###")
data = {'A': np.random.randint(1, 300, 30), 'B': np.random.randint(1, 300, 30), 'C': np.random.randint(1, 300, 30), 'D': np.random.randint(1, 300, 30)}
df = pd.DataFrame(data)
print(df.describe())
print(df.head(15))
print(df.dtypes)
print(df.isnull().sum())
print(df.corr())
print(df.sort_values(by='A', ascending=False))
print(df.sample(5))

print("\n### Experiment 3: Basic Plots using Matplotlib ###")
plt.figure(figsize=(10,6))
plt.plot(arr, arr**2, marker='o', linestyle='--', color='b', label='y=x^2')
plt.plot(arr, arr**3, marker='s', linestyle='-', color='r', label='y=x^3')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic Plot')
plt.legend()
plt.grid()
plt.show()

print("\n### Experiment 4: Frequency Distributions ###")
plt.figure(figsize=(10,6))
plt.hist(df['A'], bins=15, color='r', alpha=0.7, label='A')
plt.hist(df['B'], bins=15, color='b', alpha=0.5, label='B')
plt.title('Frequency Distribution of Columns A and B')
plt.legend()
plt.show()

print("\n### Experiment 5: Averages ###")
print(np.mean(df['A']), np.median(df['A']), stats.mode(df['A'], keepdims=True).mode[0])
print(df.mean())
print(df.median())
print(df.mode())

print("\n### Experiment 6: Variability ###")
print(np.std(df['A']), np.var(df['A']), np.max(df['A']) - np.min(df['A']))
print("Variance:\n", df.var())
print(df.std())
print(df.describe().loc[['std']])

print("\n### Experiment 7: Normal Curves ###")
x = np.linspace(-5, 5, 300)
y = stats.norm.pdf(x, 0, 1)
plt.plot(x, y, color='g', label='Standard Normal Curve')
y2 = stats.norm.pdf(x, 1, 2)
plt.plot(x, y2, color='r', linestyle='dashed', label='Mean=1, Std=2')
plt.title('Normal Distribution Curves')
plt.legend()
plt.show()

print("\n### Experiment 8: Correlation and Scatter Plots ###")
plt.figure(figsize=(10,6))
plt.scatter(df['A'], df['B'], c=df['C'], cmap='coolwarm', alpha=0.7)
plt.colorbar(label='C values')
plt.xlabel('A')
plt.ylabel('B')
plt.title('Scatter Plot between A and B')
plt.show()

print("\n### Experiment 9: Correlation Coefficient ###")
corr_matrix = df.corr()
print(corr_matrix)
print(df['A'].corr(df['B']))
plt.figure(figsize=(8,6))
plt.imshow(corr_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(corr_matrix)), corr_matrix.columns, rotation=45)
plt.yticks(range(len(corr_matrix)), corr_matrix.columns)
plt.title('Correlation Matrix Heatmap')
plt.show()

print("\n### Experiment 10: Regression ###")
slope, intercept, r_value, p_value, std_err = stats.linregress(df['A'], df['B'])
print(f"Y = {slope:.2f}X + {intercept:.2f}")
plt.figure(figsize=(10,6))
plt.scatter(df['A'], df['B'])
plt.plot(df['A'], slope*df['A'] + intercept, color='r')
plt.xlabel('A')
plt.ylabel('B')
plt.title('Linear Regression')
plt.legend()
plt.show()
```
